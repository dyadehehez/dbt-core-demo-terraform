```redshift
CREATE TABLE transaction (
    credit_card VARCHAR(18),
    date datetime,
    transaction_dollar_amount DECIMAL(37,2),
    long DECIMAL(37,13),
    lat DECIMAL(37,13)
);
```

```redshift
-- selecting only number columns both tables
SELECT c.credit_card, t.date, transaction_dollar_amount, credit_card_limit FROM transaction t
LEFT JOIN cc_info c
USING (credit_card)
LIMIT 5
```

```redshift
-- answering the question what is the largest amount per state
WITH joined AS(
SELECT credit_card_limit, state FROM transaction t
LEFT JOIN cc_info c
USING (credit_card)
)
SELECT state, MAX(credit_card_limit) AS max_credit_limit_of_state FROM joined
GROUP BY state
ORDER BY max_credit_limit_of_state DESC
LIMIT 5
```
